```python
import numpy as np
import matplotlib.pyplot as plt

sectores = ['Transporte', 'Inclusión Social Y Reconciliación', 'Salud Y Protección Social', 'Vivienda Ciudad Y Territorio', 'Educación', 'Gobierno Territorial', 'Agricultura Y Desarrollo Rural', 'Ambiente Y Desarrollo Sostenible', 'Deporte Y Recreación', 'Cultura', 'Ciencia Tecnología E Innovación', 'Minas Y Energía', 'Comercio Industria Y Turismo', 'Planeación', 'Justicia Y Del Derecho', 'Tecnologías De La Información Y Las Comunicaciones', 'Trabajo', 'Interior', 'Información Estadística', 'Hacienda', 'Presidencia De La República', 'Empleo Público', 'Organismos De Control', 'Fiscalía', 'Relaciones Exteriores', 'Rama Judicial', 'Registraduría', 'Congreso De La República', 'Sistema Integral De Verdad Justicia Reparación Y No Repetición']
inversiones = [44369, 37277, 35557, 35329, 35243, 28358, 23933, 21367, 20572, 19035, 10780, 9913, 9808, 6813, 6324, 5478, 5440, 5098, 2341, 2294, 2146, 1973, 1174, 414, 329, 267, 196, 115, 88]

log_inversiones = np.log(inversiones)

coefficients = np.polyfit(range(len(sectores)), log_inversiones, 1)
slope = coefficients[0]
intercept = coefficients[1]

plt.scatter(range(len(sectores)), log_inversiones, color='blue', label='Datos')
plt.plot(range(len(sectores)), slope * range(len(sectores)) + intercept, color='red', label='Regresión lineal')
plt.xlabel('Sectores')
plt.ylabel('log(Inversiones)')
plt.xticks(range(len(sectores)), sectores, rotation=90)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo en una tabla con valores
sectores = ['Transporte', 'Inclusión Social Y Reconciliación', 'Salud Y Protección Social', 'Vivienda Ciudad Y Territorio', 'Educación', 'Gobierno Territorial', 'Agricultura Y Desarrollo Rural', 'Ambiente Y Desarrollo Sostenible', 'Deporte Y Recreación', 'Cultura', 'Ciencia Tecnología E Innovación', 'Minas Y Energía', 'Comercio Industria Y Turismo', 'Planeación', 'Justicia Y Del Derecho', 'Tecnologías De La Información Y Las Comunicaciones', 'Trabajo', 'Interior', 'Información Estadística', 'Hacienda', 'Presidencia De La República', 'Empleo Público', 'Organismos De Control', 'Fiscalía', 'Relaciones Exteriores', 'Rama Judicial', 'Registraduría', 'Congreso De La República', 'Sistema Integral De Verdad Justicia Reparación Y No Repetición']
inversiones = [44369, 37277, 35557, 35329, 35243, 28358, 23933, 21367, 20572, 19035, 10780, 9913, 9808, 6813, 6324, 5478, 5440, 5098, 2341, 2294, 2146, 1973, 1174, 414, 329, 267, 196, 115, 88]

# Transformar los datos en formato adecuado para el modelo
X = np.arange(len(sectores)).flatten()
y = np.array(inversiones)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X.reshape(-1, 1), y)

# Realizar predicciones para nuevos valores
X_pred = np.arange(len(sectores)).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X_pred, y_pred, color='red', label='Regresión lineal')
plt.xlabel('Sectores')
plt.ylabel('Inversiones')
plt.xticks(X, sectores, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

plt.savefig("grafico_regresion.png", dpi=300, bbox_inches='tight')
```
